```python
import requests
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
h3_url = "https://comtrade.un.org/data/cache/classificationH3.json"
h4_url = "https://comtrade.un.org/data/cache/classificationH4.json"

h3 = requests.get(h3_url)
h4 = requests.get(h4_url)
```

```python
a = pd.io.json.json_normalize(h3.json()["results"])
b = pd.io.json.json_normalize(h4.json()["results"])
m = a.merge(b,on="id",how="outer")
```

```python
def fuzzy_string_index(left_miss,right_miss):
    """Creates a merge index based on the string descriptions of the goods"""
    stopwords = ["-","other","or","and"]
    for word in stopwords:
        right_miss.text_y = right_miss.text_y.str.replace(word,"") 
    choices = list(right_miss.text_y)
    #left_miss.text_x.map(process.e)
    extract = left_miss.apply(lambda row: process.extractOne(row['text_x'],choices),axis=1)
    left_miss['match_key'] = extract.map(lambda x: x[0])
    left_miss['match_score'] = extract.map(lambda x: x[1])
    return left_miss

def fuzzy_id_index(left_miss,right_miss):
    """Creates a merge index based on the difference in the commodity codes"""
    left_miss['str_id'] = left_miss.id.map(str)
    right_miss['str_id'] = right_miss.id.map(str)
    choices = right_miss.str_id
    extract = left_miss.apply(lambda row: process.extractOne(row['text_x'],choices),axis=1)
    left_miss['merge_id'] = extract.map(lambda x: x[0])
    left_miss['merge_id_score'] = extract.map(lambda x: x[1])
    left_miss['merge_id_text'] = extract.map(lambda x: right_miss[right_miss.str_id==x[0]].text_y)
    return left_miss
```

It appears the fuzzy string match works much better than fuzzy merge on the id.
